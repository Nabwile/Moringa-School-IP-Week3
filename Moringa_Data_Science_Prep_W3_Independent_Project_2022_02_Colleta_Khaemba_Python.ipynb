## Basic operations for cleaning data

```python
# Changing the columns names to be similar in the three telecom datasets
# Ensure proper spelling for ease of reading the column names
# preview the column names
# change to lower case
# This will enable it to later merge into one dataset


df1.columns
  
df1.columns = map(str.lower, df1.columns)
df1.columns

columns = ['product', 'value', 'date_time', 'cell_on_site', 'dw_a_number_int',
'dw_b_number_int', 'country_a', 'country_b', 'cell_id', 'site_id']
```

## Merging Dataframes

```python
# Merge the datasets with similar information
# I used .tail to show me the last 5 rows of the dataset 

df_combined = pd.concat([df1,df2,df3], ignore_index = 1)
df_combined.tail()
```

## Cleaning data

### Merged dataset cleaning

```python
# Dropping unnecessary columns 
# This columns are indicated in the CDR description file

df_combined1 = df_combined.drop(columns = ['country_a' ,	'country_b' ,'cell_on_site'])
df_combined1.head()
```

```python
# To check for missing value in data

df_combined1.isnull()

# This returns a boolean value for all cells that have NaN value
```

```python
#drop missing Observations in data

df_combined1_no_missing = df_combined1.dropna()
df_combined1_no_missing
```

```python
# Drop rows where all cells in that row are NA
df_combined1 = df_combined1.dropna(how='all')
df_combined1
```

```python
# To get the number of row and columns in the dataset
df_combined1.shape
```

```python
# Create 2 new columns so as to be able to separate date and time 

df_combined1[['date', 'time']] = df_combined1.date_time.str.split(expand=True)
df_combined1.head()
```

```python
# Identify which observations are duplicates
df_combined1.duplicated()

# Drop duplicates
df_combined1.drop_duplicates()
```
